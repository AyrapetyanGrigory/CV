### 1. Написать функцию на Python

#### Нужно написать функцию, которая удалит дубликаты в списке. Список не отсортирован. Необходимо сохранить порядок сортировки оригинального списка:



Для начала, создадим список, с которым будем работать:

```python
dataList = [1, 3, 2, 1, 5, 3, 5, 1, 4] 
```

Затем создадим словарь, ключами в котором будут значения из списка. Словарь нам нужен, чтобы убрать все повторяющиеся значения из списка, так как в словаре не может быть повторяющихся ключей:

```python
dict.fromkeys(dataList)
```

Далее, превратим получившийся словарь обратно в список:

```python
list(dict.fromkeys(dataList))
```

Теперь соберем все получившиеся действия в одну строчку и выведем результат:

```python
dataList = list(dict.fromkeys(dataList))
print(dataList)
```

Асимптотическая сложность этой функции равна O(n2), так как изначально сказано, что список неотсортирован, и dict.fromkeys() проходит по неотсортированной части, а list() уже проходит по отсортированной части из dict.fromkeys()
